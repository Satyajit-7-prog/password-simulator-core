```python
import pandas as pd
import joblib
from sklearn.ensemble import RandomForestClassifier
import os

# 1. Create a simple educational dataset
data = {
    'password': ['password', '123456', 'admin', 'qwerty', 'MySuperSecret99!', 'Hello@123', 'Weak1', 'Stronger#Pass2026', 'a', 'abcd', 'P@ssw0rd2024!'],
    'strength': [0, 0, 0, 0, 3, 2, 1, 4, 0, 0, 4] # 0=Terrible, 4=Unbreakable
}
df = pd.DataFrame(data)

# 2. Feature Engineering (Teaching the AI what patterns to look for)
def extract_features(pwd):
    return {
        'length': len(pwd),
        'num_digits': sum(c.isdigit() for c in pwd),
        'num_upper': sum(c.isupper() for c in pwd),
        'num_special': sum(not c.isalnum() for c in pwd)
    }

# Convert the passwords into data the AI can understand
features = pd.DataFrame(df['password'].apply(extract_features).tolist())
X = features
y = df['strength']

# 3. Train the Machine Learning Model
model = RandomForestClassifier(n_estimators=10, random_state=42)
model.fit(X, y)

# 4. Save the AI "Brain" to your models folder
# Ensure the models directory exists
os.makedirs('../models', exist_ok=True)
joblib.dump(model, '../models/password_ai_model.pkl')

print("✅ AI Model successfully trained and saved as password_ai_model.pkl!")
```
